```python
# Contador de palavras
import os
file = open('datasource/paragrafo1.txt', 'r', encoding='utf-8')
texto = file.read()
#print(texto)
```

```python
# Limpando o texto
import re
texto = re.sub('[^A-Za-z çáéíóúâôêãõ]', '', texto)

# Tokenizar o texto
from nltk.tokenize import word_tokenize
tokenized_text = word_tokenize(texto)
tokenized_text
```

```python
dic = {}
for p in tokenized_text:
    if p in dic:
        dic[p] =  dic[p] + 1
    else:
        dic[p] = 1
```

```python
%%time
lista = []
for e in dic:
    if dic[e] > 9:
        valor = str(dic[e]) + '-' + e
        lista.append(valor)
    #print(e)
#dic.items()
lista.sort()
lista.reverse()
print(lista)
```

```python
%%time
dicionario = {}
for p in lista:
    count, word = p.split('-')
    dicionario[word] = count
    
dicionario
```
